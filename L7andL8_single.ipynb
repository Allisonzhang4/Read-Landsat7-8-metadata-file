```python
import numpy as np
import netCDF4 as nc
import pandas as pd
import numpy as np
import csv
import os


def loadimage(path):
    file_obj = nc.Dataset(path, 'r', Format = 'Netcdf4')
    print(file_obj.variables.keys())
    data_mat = list()
    for i in file_obj.variables.keys():
        data_mat.append(file_obj.variables[i][:])
        
    return data_mat, file_obj


#gives the closest location of the region of interest point
def roi_location(data_mat, longi, lati):
    lon_ori, lat_ori = np.array(data_mat[0]), np.array(data_mat[1])
    new_lon, new_lat = abs(lon_ori - longi), abs(lat_ori - lati)
    
    difference = new_lon + new_lat
    result_diff = np.where(difference == np.min(difference))
    assert len(result_diff) >= 2
    assert len(result_diff) % 2 == 0
    
    final_loc = [r[0] for r in result_diff]
    
    return final_loc


#def list1(local_id, file_obj):
#    for keyid in local_id:
#        listn = []
#        for i in file_obj.variables.keys():
#            listn.append(file_obj.variables[i][tuple(local_id[keyid])].data[()])
#        
#        listn.insert(0, keyid)
#        rows.append(listn)
#    return rows


#extract the interest point variables' values
def list1(keyid, local_id, file_obj):
    listn = []
    for i in file_obj.variables.keys():
        listn.append(file_obj.variables[i][tuple(local_id[keyid])].data[()])
    return listn


def main(aim_lat, aim_lon, local_id, image_filepath, save_folder):
    aim = [x for x in zip(aim_lon, aim_lat)]

    data_mat, file_obj = loadimage(image_filepath) 

    roi_list = {}
    for i in aim:
        roi_list[i] = roi_location(data_mat, i[0], i[1])
        
    rows = []
    for i in local_id:
        li1 = list1(i, local_id, file_obj)
        li1.insert(0, i)
        rows.append(li1)

    if save_folder:
        if not os.path.exists(save_folder):
            os.makedirs(save_folder)
        save_filepath = os.path.join(save_folder, image_filepath.split('/')[-1].replace('nc', 'csv'))
    else:
        save_filepath = image_filepath.replace('nc', 'csv')
    
    with open(save_filepath, 'w', newline='') as f:
        fwr = csv.writer(f)

        firstrow = ['idloc']
        firstrow.extend(list(file_obj.variables.keys()))
        
        fwr.writerow(firstrow)
        for row in rows:
            fwr.writerow(row)
```
